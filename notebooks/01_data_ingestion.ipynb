## Step 2 — Load & Inspect Raw Data

```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python

df = pd.read_csv(r'data\healthcare_dataset.csv')
df.shape
df.head()
```

```python
df.info()
```

```python
# check for misssing values
df.isnull().sum()
```

```python
df.shape
```

```python
# duplicates
df.duplicated().any()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.duplicated().sum()
```

```python
# change the date format
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

```python
df.info()
```

```python
# standardize categorical columns
df['Gender'] = df['Gender'].str.capitalize()
```

```python
df["Insurance Provider"].value_counts()
```

```python
# drop columns
df.drop(columns=[ 'Name',"Room Number","Insurance Provider","Discharge Date",'Date of Admission',"Doctor","Hospital"], inplace=True)
```

```python
# save the cleaned data
df.to_csv(r'C:\Users\user\OneDrive\Desktop\HEALTH\data\cleaned_data.csv', index=False)
```
